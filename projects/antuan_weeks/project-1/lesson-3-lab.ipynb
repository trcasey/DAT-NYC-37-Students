```python
import pandas as pd
import os

data = pd.read_csv(os.path.join('../../..', 'DAT-NYC-37', 'lessons', 'lesson-02', 'assets', 'dataset', 'ozone.csv'))
```

```python
data.columns.values
```

Print the column names of the dataset to the screen, one column name per line.  

```python
for column in data:
    print column
```

Extract the first 2 rows of the data frame and print them to the console. What does the output
look like?

```python
data.iloc[:2] # or data.head(2)
```

How many observations (i.e. rows) are in this data frame?

```python
len(data) # returns number of rows
```

Extract the last 2 rows of the data frame and print them to the console. What does the output
look like?

```python
data[-2:] # or data.tail(2)
```

What is the value of Ozone in the 47th row?

```python
data.loc[[46], ['Ozone']] # viewing 46th index with indexes beginning at 0 gives the 47th row
```

How many missing values are in the Ozone column of this data frame?

```python
len(data) - data['Ozone'].count() # total rows - rows with Ozone data OR data['Ozone'].isnull().sum()
```

What is the mean of the Ozone column in this dataset? Exclude missing values (coded as NA)
from this calculation.

```python
data['Ozone'].mean(skipna=True) # data.Ozone returns series instead of dataframe object; series don't have columns
```

Extract the subset of rows of the data frame where Ozone values are above 31 and Temp values
are above 90. What is the mean of Solar.R in this subset?

```python
data[(data['Ozone'] > 31) & (data['Temp'] > 90)]['Solar.R'].mean() # mean of Solar.R where Ozone > 31 & Temp > 90
```

What is the mean of "Temp" when "Month" is equal to 6?

```python
data[data['Month'] == 6]['Temp'].mean() # mean of temp where Month == 6
```

What was the maximum ozone value in the month of May (i.e. Month = 5)?

```python
data[data['Month'] == 5]['Ozone'].max() # maximum value where Month == 5
```
